```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# annual percentage rate (APR) = r
# effective annual rate = (1+r/n)^n - 1
# where r/n is per-period rate for each period, and n is period of compounding
# EAR gives the total amount of interest you'd expect to get if you leave money in your savings account for a year
```

```python
# what will effective annual rate look like when n approaches infinity, given annual percentage rate = r?
def ear(r, n):
    return (1 + r/n)**n - 1
```

```python
compounding_periods_list = list(np.linspace(1,1000,1000))
ear_list = [ear(0.05, n) for n in compounding_periods_list]
```

```python
plt.plot(compounding_periods_list, ear_list)
plt.axhline(y = np.exp(0.05) - 1.0, color = 'r', linestyle = '--')
plt.xlabel('Compounding Periods')
plt.ylabel('Effective Annual Rate')
plt.title('Effective Annual Rate as a function of compounding periods, APR = 0.05') 
plt.text(20, 0.0512, 'limit value $= e^{0.05} - 1 = 05127...$', fontsize = 10)
plt.show()

# (1 + x/n)^n -> e^x as n -> infinity
# therefore ear(x, n) -> e^x - 1 as n -> infinity
# Given APR = x, if you have continous compounding, you'd expect to earn (e^x - 1) as interest
```
