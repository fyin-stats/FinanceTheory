```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mortage interest rate is a funciton of the following inputs
# (yearly) APR (assuming monthly compounding)
# home_value
# downpayment_value
# terms: number of payments

# the calculation is based on the below equation
# home_value - downpayment = PV = discounted cashflow
# where discounted_cashflow = cf1/(1+r) + ... + cfn/(1+r)^n
# where n is the number of payments
# and r is calculated from APR/12
```

```python
def monthly_payment_calculator(home_value, downpayment_value, terms, apr):
    pv = home_value - downpayment_value
    r = apr / 12
    # C * (1/(1+r) + 1/(1+r)**2 + ...) = PV
    return pv/sum([1/(1+r)**x for x in range(1, terms + 1)])
```

```python
monthly_payment_calculator(2800000, 800000, 360, 0.0625)
```

```python
# Fixing total loan value, vary apr to take values between 2.5% to 10.0%, can we get a curve of monthly payments?
apr_list = np.linspace(0.02, 0.1, 100)
# print(apr_list)

monthly_payment_list = [monthly_payment_calculator(2000000, 800000, 360, x) for x in apr_list]
plt.plot(apr_list, monthly_payment_list)
plt.title("Monthly payment as a function of APR (decimal), given $1200000 loan value")
plt.show()
```
